```python
latitude_degrees = 10.618
latitude_radians = latitude_degrees * math.pi / 180
latitude_radians
```

```python
J = np.arange(1, 366)  # Days from 1 to 365

phi = 10.618 * np.pi / 180  # Convert latitude to radians
phi
```

```python
 # Calculation of dr, delta, omega_s
d_r = 1 + 0.033 * np.cos((2 * np.pi / 365) * J)

delta = 0.409 * np.sin((2 * np.pi / 365) * J - 1.39)

def omega_s(phi, delta):
     return np.arccos(-np.tan(phi) * np.tan(delta))
```

```python
# # Calculation of Day_length
N = (2 / 15) * omega_s(phi, delta)
```

```python
# # Calculation of R_a (Extraterrestrial Solar Flux)
H_o = (24 * 60 / np.pi) * 0.0820 * d_r * (omega_s(phi, delta) * np.sin(phi) * np.sin(delta) + 
                                               np.cos(phi) * np.cos(delta) * np.sin(omega_s(phi, delta)))
```
